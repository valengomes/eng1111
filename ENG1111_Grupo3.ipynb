```python
import pandas as pd
import numpy
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt

# Importando o database de notas de matematica
student_data_mat=pd.read_csv('student-mat.csv',sep=';')


# Limpando dados
student_data_mat.dropna()
student_data_mat = student_data_mat.drop('school', axis=1)
student_data_mat['internet'] = student_data_mat['internet'].map( {'yes': 1, 'no': 0})
student_data_mat['schoolsup'] = student_data_mat['schoolsup'].map( {'yes': 1, 'no': 0})
student_data_mat['paid'] = student_data_mat['paid'].map( {'yes': 1, 'no': 0})
student_data_mat['activities'] = student_data_mat['activities'].map( {'yes': 1, 'no': 0})
student_data_mat['nursery'] = student_data_mat['nursery'].map( {'yes': 1, 'no': 0})
student_data_mat['higher'] = student_data_mat['higher'].map( {'yes': 1, 'no': 0})
student_data_mat['romantic'] = student_data_mat['romantic'].map( {'yes': 1, 'no': 0})
student_data_mat['Female'] = student_data_mat['sex'].map({'F':1,'M':0}) 
student_data_mat['Male'] = student_data_mat['sex'].map({'M':1,'F':0}) 
student_data_mat['Urban'] = student_data_mat['address'].map({'U':1,'R':0}) 
student_data_mat['Rural'] = student_data_mat['address'].map({'R':1,'U':0}) 
student_data_mat['SmallFam'] = student_data_mat['famsize'].map({'LE3':1,'GT3':0}) 
student_data_mat['BigFam'] = student_data_mat['famsize'].map({'GT3':1,'LE3':0}) 
student_data_mat['Mteacher'] = student_data_mat['Mjob'].map({'teacher':1,'health':0,'services':0,'other':0,'at_home':0}) 
student_data_mat['Mhealth'] = student_data_mat['Mjob'].map({'health':1,'teacher':0,'services':0,'other':0,'at_home':0}) 
student_data_mat['Mservices'] = student_data_mat['Mjob'].map({'services':1,'teacher':0,'health':0,'other':0,'at_home':0}) 
student_data_mat['M_other'] = student_data_mat['Mjob'].map({'other':1,'teacher':0,'services':0,'health':0,'at_home':0}) 
student_data_mat['M_at_home'] = student_data_mat['Mjob'].map({'at_home':1,'teacher':0,'services':0,'health':0,'other':0}) 
student_data_mat['Fteacher'] = student_data_mat['Fjob'].map({'teacher':1,'health':0,'services':0,'other':0,'at_home':0}) 
student_data_mat['Fhealth'] = student_data_mat['Fjob'].map({'health':1,'teacher':0,'services':0,'other':0,'at_home':0}) 
student_data_mat['Fservices'] = student_data_mat['Fjob'].map({'services':1,'teacher':0,'health':0,'other':0,'at_home':0}) 
student_data_mat['F_other'] = student_data_mat['Fjob'].map({'other':1,'teacher':0,'services':0,'health':0,'at_home':0}) 
student_data_mat['F_at_home'] = student_data_mat['Fjob'].map({'at_home':1,'teacher':0,'services':0,'health':0,'other':0}) 
student_data_mat['Rhome'] = student_data_mat['reason'].map( {'home': 1, 'reputation': 0,'course':0,'other':0})
student_data_mat['Rreputation'] = student_data_mat['reason'].map( {'reputation': 1, 'home': 0,'course':0,'other':0})
student_data_mat['Rcourse'] = student_data_mat['reason'].map( {'course': 1, 'reputation': 0,'home':0,'other':0})
student_data_mat['Rother'] = student_data_mat['reason'].map( {'other': 1, 'reputation': 0,'course':0,'home':0})
student_data_mat['Gmother'] = student_data_mat['guardian'].map( {'mother': 1, 'father': 0,'other':0})
student_data_mat['Gfather'] = student_data_mat['guardian'].map( {'father': 1, 'mother': 0,'other':0})
student_data_mat['Gother'] = student_data_mat['guardian'].map( {'other': 1, 'mother': 0,'father':0})
student_data_mat['Ptogether'] = student_data_mat['Pstatus'].map({'T':1,'A':0}) 
student_data_mat['Papart'] = student_data_mat['Pstatus'].map({'A':1,'T':0}) 
student_data_mat = student_data_mat.select_dtypes(exclude=['object'])

print(student_data_mat.failures)

# Criando o HEATMAP do database de matematica
plt.figure(figsize=(40,30))
heatmap = sns.heatmap(student_data_mat.corr(),vmin= -1, vmax = 1, annot = True,fmt=".3f",cmap='PiYG')


student_data_mat.dtypes
```
